```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy as sc
import os
```

```python
if not os.path.exists('output_figures'):
    os.mkdir('output_figures')
```

```python
data1 = pd.read_excel('Dataset_muSORS.xlsx', sheet_name='Fig. 2a')
```

```python
bmi = data1['BMI']
age = data1['Age']
```

```python
sns.set_theme()
sns.set_style('white')
```

```python
fig, axes = plt.subplots(figsize=(5,5), ncols=2, sharey=True)
fig.tight_layout()
axes[0].barh(np.arange(35)+0.5, age, edgecolor='none', facecolor='b')
axes[1].barh(np.arange(35)+0.5, bmi, edgecolor='none', facecolor='r')
axes[0].invert_xaxis()
axes[0].invert_yaxis()
axes[1].invert_yaxis()
axes[0].set_xticks(np.arange(9)*10)
axes[1].set_xticks(np.arange(5)*10)
axes[0].set_xlim([0,80])
axes[1].set_xlim([0,40])
axes[0].set_ylim([0,35])
axes[1].set_ylim([0,35])
axes[0].invert_xaxis()
axes[0].invert_yaxis()
plt.savefig('output_figures/age_bmi.pdf')
```

```python
data2 = pd.read_excel('Dataset_muSORS.xlsx', sheet_name='Fig. 2b')
```

```python
VPG = data2.to_numpy()[3:, 1:].T.astype('float')
```

```python
vmin = np.min(VPG)
vmax = np.max(VPG)
```

```python
plt.imshow(VPG[:20,:], aspect=1/3, cmap='rainbow', interpolation='none', vmin=vmin, vmax=vmax)
plt.xticks(np.arange(12))
plt.colorbar()
plt.savefig('output_figures/VPG_d.pdf')
```

```python
plt.imshow(VPG[20:,:], aspect=1/3, cmap='rainbow', interpolation='none', vmin=vmin, vmax=vmax)
plt.xticks(np.arange(12))
plt.colorbar()
plt.savefig('output_figures/VPG_n.pdf')
```
