```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
if not os.path.exists('output_figures'):
    os.mkdir('output_figures')
```

```python
data = pd.read_excel('Dataset_muSORS.xlsx', sheet_name='Fig. 2g')
```

```python
data = data.to_numpy()
data
```

```python
sns.set_theme()
sns.set_style('white')
sns.set_style('ticks')
color_list = sns.color_palette('tab10')
```

```python
for i in range(5):
    plt.scatter(data[13,1+i], data[12,1+i], color=color_list[i])
plt.savefig('output_figures/Slope_CORR.pdf')
```

```python
for i in range(5):
    plt.scatter(data[:10,0], data[:10,i+1], color=color_list[i])
    plt.plot(data[:10,0], data[12,1+i]*data[:10,0]+data[14,1+i], color=color_list[i], linestyle='dashed')
plt.savefig('output_figures/VPG_spectra.pdf')
```
