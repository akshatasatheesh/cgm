```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import metrics
import seaborn as sns
import scipy.stats as st
from matplotlib import colors
```

```python
if not os.path.exists('output_figures'):
    os.mkdir('output_figures')
```

```python
data = pd.read_excel('Dataset_muSORS.xlsx', sheet_name='Extended Data Fig. 10')
data = data.to_numpy()
```

```python
time1 = data[6:, 1]
rmse1 = data[6:, 2]
```

```python
sns.set_theme()
sns.set_style('white')
sns.set_style('ticks')
plt.plot(time1, rmse1, marker='o', color=sns.color_palette()[3])
plt.savefig('output_figures/preliminary_offset3_backtrack.pdf')
```

```python
time2 = data[1:27, 5]
offset2 = data[1:27, 6]
offset3 = data[1:27, 7]
offset23 = data[1:27, 8]
```

```python
plt.plot(time2, offset2, marker='o', color=sns.color_palette()[2])
plt.plot(time2, offset3, marker='o', color=sns.color_palette()[3])
plt.plot(time2, offset23, marker='o', color='k')
plt.savefig('output_figures/expanded_cross_backtrack.pdf')
```
